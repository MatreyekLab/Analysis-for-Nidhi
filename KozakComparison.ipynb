```python
# This script finds the frequency of every kozak sequence that is shared between the libary 
# and recombined cells, as well as the template sequence percentage

import pandas as pd
pd.set_option('display.max_rows', None)

library = pd.read_csv("Hyg_kz_plasmid.assembled.tsv", sep='\t')
recombine = pd.read_csv("Hyg_KZ.assembled.tsv", sep='\t')
template_library = library['kozak_sequence'].isna().sum()/len(library.index)
template_recombine = recombine['kozak_sequence'].isna().sum()/len(recombine.index)
data = [['Library',template_library], ['Recombine',template_recombine]]
template = pd.DataFrame(data, columns = ['Template', 'Percentage'])
template
```

```python
lib = library.groupby('kozak_sequence',as_index=False).size()
lib['Library_%'] = (lib['size']*100)/len(library.index)
lib.rename(columns={'size':'Number_library'}, inplace=True)
rec = recombine.groupby('kozak_sequence',as_index=False).size()
rec['Recombine_%'] = (rec['size']*100)/len(recombine.index)
rec.rename(columns={'size':'Number_recombine'}, inplace=True)
common = pd.merge(lib,rec,how="inner",on='kozak_sequence')
common
```
